**Questão número 10 - Elabore um gráfico de linhas com a quantidade de reclamações por mês. Este gráfico lembra alguma distribuição estatística?**

```python

df_consults = dfs
```

Feito isso, testaremos!

Uma vez que deu tudo certo, precisamos descobrir na coluna qual tipo de dado a colulna 'DataAbertura' representa. 

```python
df_consults.shape
```

Uma vez que descobrimos que se trata de uma string, logo, para conseguirmos realizar corretamente a tabela, precisamos converter os dados para dados de data:

```python
df_consults['DataAbertura'] = pd.to_datetime(df_consults['DataAbertura'])
```

Feito isso, faremos um teste para saber se a conversão foi realizada corretamente.

```python
type(df_consults['DataAbertura'][0])
```

```python
df_consults['DataAbertura'].head()
```

Agora iremos criar os meses no dataframe, o que facilitará a construção correta do gráfico em questão:

```python
df_consults['Mes Abertura'] = df_consults['DataAbertura'].dt.month
```

Agora iremos consultar os dados, mas dessa vez convertidos.

```python
df_consults['Mes Abertura'].head()
```

Agora iremos realizar a contabilidade de reclamações por mês no ano de 2016:

```python
df_consults['Mes Abertura'].value_counts()
```

Agora, antes de gerar o gráfico, precisamos criar um dataFrame intermediário:

```python
dados = df_consults['Mes Abertura'].value_counts()
```

Agora iremos criar o gráfico em linhas, conforme solicitado na atividade:

```python
sns.lineplot(dados.index, dados.values, alpha=0.8)
plt.title('Número de Reclamações por Mês')
plt.ylabel('Numero de Reclamações')
plt.xlabel('Meses', fontsize=14)
```
